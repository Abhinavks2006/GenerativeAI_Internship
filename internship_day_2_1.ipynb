```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/preprocessing_practice_dataset.csv")
```

```python
df['Age_mean_imputed']=df['Age'].fillna(df['Age'].mean())
df['Age_median_imputed']=df['Age'].fillna(df['Age'].median())
df['Department_mode_imputed']=df['Department'].fillna(df['Department'].mode()[0])
df['Education_mode_imputed']=df['Education'].fillna(df['Education'].mode()[0])
df['Monthly_Salary_mean_imputed']=df['Monthly_Salary'].fillna(df['Monthly_Salary'].mean())
df['Monthly_Salary_median_imputed']=df['Monthly_Salary'].fillna(df['Monthly_Salary'].median())
df['City_mode_imputed']=df['City'].fillna(df['City'].mode()[0])
df[['Age','Age_mean_imputed','Age_median_imputed','Department','Department_mode_imputed','Education','Education_mode_imputed','Monthly_Salary','Monthly_Salary_mean_imputed','Monthly_Salary_median_imputed','City','City_mode_imputed']]
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10,6))
sns.boxplot(data=df,y='Age_mean_imputed',x='Department_mode_imputed',palette='pastel')
plt.title('Boxplot of Department_mode_imputed by Age Class',fontsize=16)
plt.ylabel('Age Class',fontsize=12)
plt.xlabel('Department_mode_imputed',fontsize=12)
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10,6))
sns.boxplot(data=df,y='Age_mean_imputed',x='Education_mode_imputed',palette='magma')
plt.title('Boxplot of Education_mode_imputed by Age Class',fontsize=16)
plt.ylabel('Age Class',fontsize=12)
plt.xlabel('Education_mode_imputed',fontsize=12)
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Department_mode_imputed',y='Monthly_Salary_mean_imputed',palette='magma')
plt.title('Boxplot of 	Monthly_Salary_mean_imputed by Department Class',fontsize=16)
plt.xlabel('Department Class',fontsize=12)
plt.ylabel('Monthly_Salary_mean_imputed',fontsize=12)
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Gender',y='Department_mode_imputed',palette='pastel')
plt.title('Boxplot of Gender by Department_mode_Class',fontsize=16)
plt.ylabel('Department_mode_Class',fontsize=12)
plt.xlabel('Gender',fontsize=12)
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Department_mode_imputed',y='City_mode_imputed',palette='pastel')
plt.title('Boxplot of Department_mode_imputed by City_mode_imputed',fontsize=16)
plt.xlabel('Department_mode_imputed',fontsize=12)
plt.ylabel('City_mode_imputed',fontsize=12)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Distribution of Age (Original vs. Mean Imputed)
sns.histplot(df['Age'], color='blue', label='Original Age', kde=True, ax=axes[0])
sns.histplot(df['Age_mean_imputed'], color='red', label='Mean Imputed Age', kde=True, ax=axes[0])
axes[0].set_title('Distribution of Age (Original vs. Mean Imputed)', fontsize=14)
axes[0].legend()

# Distribution of Monthly_Salary (Original vs. Mean Imputed)
sns.histplot(df['Monthly_Salary'], color='blue', label='Original Monthly Salary', kde=True, ax=axes[1])
sns.histplot(df['Monthly_Salary_mean_imputed'], color='red', label='Mean Imputed Monthly Salary', kde=True, ax=axes[1])
axes[1].set_title('Distribution of Monthly Salary (Original vs. Mean Imputed)', fontsize=14)
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

encoded_data=df.copy()

label_encoder=LabelEncoder()
encoded_data['Gender_Label']=label_encoder.fit_transform(encoded_data['Gender'])

one_hot_encoded=pd.get_dummies(encoded_data['Department_mode_imputed'],prefix='Department')
encoded_data=pd.concat([encoded_data,one_hot_encoded],axis=1)

perfomance_mapping = {1: 'First', 2: 'Second', 3: 'Third', 4: 'Fourth', 5: 'Five'}
encoded_data['Performance_Label'] = encoded_data['Performance_Rating'].map(perfomance_mapping)
encoded_data['Performance_Label'] = label_encoder.fit_transform(encoded_data['Performance_Label'])

print(encoded_data[['Gender','Gender_Label','Department_Finance','Department_HR','Department_IT','Department_mode_imputed']])
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
ppd_data=df.copy()
numerical_columns=['Age','Monthly_Salary','Experience_Years','Performance_Rating']
ppd_data.dropna(subset=numerical_columns,inplace=True)

min_max_scaler=MinMaxScaler()
standard_scaler=StandardScaler()

ppd_data['Age_MinMax']=min_max_scaler.fit_transform(ppd_data[['Age']])
ppd_data['Monthly_Salary_MinMax']=min_max_scaler.fit_transform(ppd_data[['Monthly_Salary']])
ppd_data['Experience_Years_MinMax']=min_max_scaler.fit_transform(ppd_data[['Experience_Years']])
ppd_data['Performance_Rating_MinMax']=min_max_scaler.fit_transform(ppd_data[['Performance_Rating']])

ppd_data['Age_Standard']=standard_scaler.fit_transform(ppd_data[['Age']])
ppd_data['Monthly_Salary_Standard']=standard_scaler.fit_transform(ppd_data[['Monthly_Salary']])
ppd_data['Experience_Years_Standard']=standard_scaler.fit_transform(ppd_data[['Experience_Years']])
ppd_data['Performance_Rating_Standard']=standard_scaler.fit_transform(ppd_data[['Performance_Rating']])

ppd_data[['Age','Age_MinMax','Age_Standard','Monthly_Salary','Monthly_Salary_MinMax','Monthly_Salary_Standard','Experience_Years','Experience_Years_MinMax','Experience_Years_Standard','Performance_Rating','Performance_Rating_MinMax','Performance_Rating_Standard']]
```
