```python
df['Age_mean_imputed']=df['Age'].fillna(df['Age'].mean())
df['Age_median_imputed']=df['Age'].fillna(df['Age'].median())
df['Cabin_mode_imputed']=df['Cabin'].fillna(df['Cabin'].mode()[0])
df[['Age','Age_mean_imputed','Age_median_imputed','Cabin','Cabin_mode_imputed']]
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Pclass',y='Age',palette='pastel')
plt.title('Boxplot of Age by Passenger Class',fontsize=16)
plt.xlabel('Passenger Class',fontsize=12)
plt.ylabel('Age',fontsize=12)
plt.show()

```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Pclass',y='Survived',palette='pastel')
plt.title('Boxplot of Survivede by Passenger Class',fontsize=16)
plt.xlabel('Passenger Class',fontsize=12)
plt.ylabel('Survived',fontsize=12)
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Embarked',y='Fare',palette='pastel')
plt.title('Boxplot of Fare by Embarked Class',fontsize=16)
plt.xlabel('Embarked Class',fontsize=12)
plt.ylabel('Fare',fontsize=12)
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
encoded_data=df.copy()
#1.label encoding for 'Sex'
label_encoder=LabelEncoder()
encoded_data['Sex_Label']=label_encoder.fit_transform(encoded_data['Sex'])
#2. One-hot Encoding for 'Embarked'
one_hot_encoded = pd.get_dummies(encoded_data['Embarked'], prefix='Embarked')
encoded_data = pd.concat([encoded_data, one_hot_encoded], axis=1)

# Since 'Pclass' is ordinal, we can create a mapped column for demonstration.
pclass_mapping = {1: 'First', 2: 'Second', 3: 'Third'}
# 3. Ordinal Encoding For p class
encoded_data['Pclass_Category'] = encoded_data['Pclass'].map(pclass_mapping)

# Display the resulting dataset with relevant encoded columns
print(encoded_data[['Sex', 'Sex_Label', 'Embarked', 'Embarked_Q', 'Embarked_S', 'Embarked_C', 'Pclass', 'Pclass_Category']].head())
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
file_path='/content/titanic uncleaned.csv'
titanic_data=pd.read_csv(file_path)

numerical_columns=['Age','Fare','Pclass']

titanic_data.dropna(subset=numerical_columns,inplace=True)

minmax_scaler=MinMaxScaler()
standard_scaler=StandardScaler()

titanic_data['Age_MinMax']=minmax_scaler.fit_transform(titanic_data[['Age']])
titanic_data['Fare_MinMax']=minmax_scaler.fit_transform(titanic_data[['Fare']])

titanic_data['Age_Standard']=standard_scaler.fit_transform(titanic_data[['Age']])
titanic_data['Fare_Standard']=standard_scaler.fit_transform(titanic_data[['Fare']])

titanic_data[['Age','Age_MinMax','Age_Standard','Fare','Fare_MinMax','Fare_Standard']]
```
